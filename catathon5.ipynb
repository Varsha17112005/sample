```python
import cv2
import torch
import numpy as np
import torchvision.transforms as transforms
import gradio as gr
from PIL import Image
from ultralytics import YOLO
```

```python
# Load RT-MDE Depth Estimation Model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
midas = torch.hub.load("intel-isl/MiDaS", "MiDaS_small").to(device)
midas.eval()
```

```python
# Load YOLOv8 Object Detection Model
yolo_model = YOLO("yolov8n.pt")  # Lightweight & fast model
```

```python
# Define Preprocessing for RT-MDE Model
transform = transforms.Compose([
    transforms.Resize((384, 384)),  # Optimized input size
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5])
])
```

```python
def estimate_depth(input_image):
    """Runs monocular depth estimation on an input image."""
    image = Image.fromarray(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
    input_tensor = transform(image).unsqueeze(0).to(device)

    with torch.no_grad():
        depth_map = midas(input_tensor).squeeze().cpu().numpy()

    depth_map = cv2.resize(depth_map, (input_image.shape[1], input_image.shape[0]))  # Match original size
    depth_map = (depth_map - np.min(depth_map)) / (np.max(depth_map) - np.min(depth_map))  # Normalize
    heat_map = cv2.applyColorMap((depth_map * 255).astype(np.uint8), cv2.COLORMAP_JET)

    return depth_map, heat_map
```

```python
def detect_objects(input_image, depth_map):
    """Runs object detection and calculates distance from POV to objects."""
    results = yolo_model(input_image)
    detected_image = input_image.copy()
    pov_x, pov_y = detected_image.shape[1] // 2, detected_image.shape[0] // 2  # Center POV

    distances = []  # Store distances

    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
            label = result.names[int(box.cls[0])]

            # Extract depth for detected object
            object_depth = np.mean(depth_map[y1:y2, x1:x2])
            if np.isnan(object_depth) or object_depth <= 0:
                estimated_distance = "N/A"
            else:
                estimated_distance = round(15 / object_depth, 2)  # Convert to meters

            # Calculate POV distance (pixel-based)
            obj_center_x, obj_center_y = (x1 + x2) // 2, (y1 + y2) // 2
            pov_distance = round(np.sqrt((pov_x - obj_center_x)**2 + (pov_y - obj_center_y)**2), 2)

            # Store distance information
            distances.append(f"{label}: {estimated_distance}m | POV Dist: {pov_distance}px")

            # Draw bounding boxes and add text
            cv2.rectangle(detected_image, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(detected_image, f"{estimated_distance}m", (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
            cv2.putText(detected_image, f"POV {pov_distance}px", (x1, y1 - 30),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

    return detected_image, "\n".join(distances)
```

```python
def process_image(input_image):
    """Processes an image and returns object detection, heatmap, and distances."""
    depth_map, heat_map = estimate_depth(input_image)
    detected_objects, distances = detect_objects(input_image, depth_map)

    return detected_objects, heat_map, distances
```

```python
# Create Gradio Interface
iface = gr.Interface(
    fn=process_image,
    inputs=gr.Image(type="numpy", label="Upload or Capture Image"),
    outputs=[
        gr.Image(type="numpy", label="Object Detection with Distance"),
        gr.Image(type="numpy", label="Depth Heat Map"),
        gr.Textbox(label="Object Distances from POV")
    ],
    live=True
)

iface.launch(debug=True)
```
